#Değer Döndüren ve Değer Döndürmeyen Fonksiyonlar

Bizim oluşturduğumuz daha önceki fonksiyonlar **değer döndürmeyen** fonksiyonlardı. Yani işlemi yaptırır ve sadece bir kod satırında çağırarak çalışırlardı. Ancak **değer döndüren** fonksiyonlar ise belirli bir tür değişkenin içerisine veri aktarılarak çalışan fonksiyonlardır.

##Peki, Değer Döndüren Fonksiyon Ne Demek Oluyor?
Değer döndürmeyen fonksiyon bir değişkenin içerisine değer atılarak yazılır. Bu değişken fonksiyonda belirlenen aynı türden değişken olmalıdır. Yani Bir fonksiyon sonucunda değişkene sadece bir değer atılır. Bu değer string ise string türünde, integer ise integer türünde bir değişkenin içine yazılır.

````python
#Değer Döndürmeyen fonksiyon
def toplama(sayi1,sayi2):
  toplama = sayi1 + sayi2
  print("Toplam:", toplama)
````

````python
#Değer Döndüren Fonksiyon
def toplam(sayi1,sayi2):
  return sayi1 + sayi2 #Return komutunu kullanıyorum.

#ANA PROGRAM
a = int(input("Sayı gir:"))
b = int(input("2.Sayıyı gir:"))
sonuc = toplam(a,b) #Bir değişken içerisine yazıyorum
print(sonuc)
````

````python
#İki sayının ortalamasını veren değer döndüren fonksiyonu yazın.
def ortalamam(sayi1,sayi2):
  return (sayi1 + sayi2) / 2

sonuc = ortalamam(65,86)
print("Sonucum: ", sonuc)
durumum = not_derecesi(sonuc)
print(durumum)
````

#ÖDEV
Bu hafta sayısal loto programı yapmanızı istiyorum. Program şöyle çalışacak:

1. Kullanıcı 6 tane 1 ile 50 arasında sayılar girecek.
2. Bilgisayar rastgele 1 ile 50 arasında 6 tane rastgele sayı üretecek.
3. Bilgisayarın ve kullanıcının girdiği rakamların her birine tek tek bakılacak ve 6 rakamın hepsi eşleşirse kullanıcıya "250 Milyon Roblox Kazandın!" denilecek.

````python
import random
#1. Kendi kağıdımı oluşturuyorum -> bir liste
kullanici_listesi = []
#2. Makinenin kağıdını oluşturuyorum -> bir liste
rastgele_listesi = []
#3.Kendi kağıdımı dolduruyorum 6 rakamla
for liste_doldurma in range(0,6):
  sayim = int(input("1-50 arasında sayı giriniz: "))
  kullanici_listesi.append(sayim)
#4.Makinenin listesini dolduruyorum.
for liste_doldurma in range(0,6):
  sayim = random.randint(1,50)
  rastgele_listesi.append(sayim)
#5.Kendi listem ve rastgele liste arasında elemanları sırayla karşılaştırma
for c in range(0,6):
  if kullanici_listesi[c] == rastgele_listesi[c]:
    print("Bir eşleşme gerçekleşti!")
    print(kullanici_listesi[c] ,"=", rastgele_listesi[c])
print(kullanici_listesi)
print(rastgele_listesi)
````
