#DONGÜLER (For ve While Döngüsü)
---
Diyelim ki, birden fazla değişkenimiz veya çok fazla kendini tekrar eden işlemimiz var. İşte o zaman döngüler kullanıyoruz. Örneğin;

1. 1'den 1000'e kadar sayıları yazdırmak istiyorsam,
2. Elimde 1000 değişkenlik bir listem var ve yazdırmak istiyorsam,
3. İstediğim bir koşul sağlandığı müddetçe çalıştıracak işlemlerim varsa,
4. Kendini tekrar eden işlemlerim varsa döngüleri kullanırım.

Toplam iki farklı döngü vardır. Biri **For**, diğeri ise **While** döngüsüdür. Şimdi **For Döngüsünü** inceleyelim.

```python
#For Döngüsünü Kurmak için yapılması gerekenler oldukça basittir:
for i in range (1,101):
  print(i)
# Her zaman son satırı almaz. O yüzden bir rakam fazla yazarım.
# 100 değil 101 yazarım.
# yukarıdaki döngüyü inceleyelim:
# for .... in ..... : bir kalıptır.
# i yazan şey bir değişkendir.
# range ise bir aralık anlamına gelmektedir.
# Range yerine başka şeyler yazılabilir.
#Bir başka örnek daha yapalım:
ismim = "Uğur Sırvermez"
for a in ismim:
  print(a)
#Adımın harflerini for döngüsü ile parçalayabilirim.
#Satır satır yazdırabilirim.
```

##For Döngüsünde İşlemler

For döngüsünde sadece sayıları art arda yazdırmayız. Aynı zamanda bir çok farklı işlemler yapabiliriz.

```python
ismim = "Ali Kerem"

for degisken in range(0,len(ismim)):
  #print(degisken)
  print(ismim)
```

```python
sayi1=int(input("Lütfen döngünün başlayacağı sayıyı yazınız:"))
sayi2=int(input("Lütfen döngünün biteceği sayıyı yazınız:"))
for degisken in range(sayi1,sayi2):
  print(degisken)
```

```python
for a in range(20,0,-2):
  print(a)
```

**For döngüsünde range() üç değer alabilir. Birincisi başlangıç, ikincisi bitiş değeri, üçüncüsü ise döngünün kaçar kaçar artacağını gösterir.**

```python
for a in range(0,10):
  if (a % 5 == 0):
    print(a)
  else:
    print("Bu sayı o sayı değil!")
```

**NOT:** **% (Yüzde)** işareti sayının modunu alma anlamına gelir. Örneğin bir sayının bölümünden kalanı verir. Mesela "a" sayısının 6'ya bölümünden kalanına mod denir.

```python
for i in range(0,50):
  if(i % 15 == 0):
    print("15'e tam bölünen sayı:",i)
  if(i % 5 == 0 or i % 3 == 0):
    print(i)
```

**YAPILACAK:** Bilgisayar rastgele iki sayı üretilecek ve üretilen sayılar arasında bir döngü kurulacak.

Birinci sayı: 0-50 arasında olacak.

İkinci sayı: 60-250 arasında olacak.

**Döngü 3'er 3'er artacak!**

```python
import random
sayi1 = random.randint(0,50)
sayi2 = random.randint(60,250)
print(sayi1," ",sayi2)

for c in range(sayi1,sayi2,3):
  print(c)
```

#İç İçe Geçen For Döngüleri Olabilir
Bazen birden fazla for döngüsü birbirinin içinde yazılabilir. Bu biraz karmaşıktır. Çok fazla pratik yapmak gerekir.

```python
for a in range (0,5):
  print("a değişkeni şunu yazdı:",a)
  for b in range(0,5):
    print("b değişkeni şunu yazdı:",b)
```
