Bazen çok sayıda tekrar eden işlem yaparız. **"Sürekli Tekrarla"** komutumuzla aynı anlama gelen işlemlere döngüler deriz. Bizim döngülerimiz bazen 0'dan 100'e kadar saymayı da kolaylıkla yapmamızı sağlar.  

İki tane döngü şeklimiz var: **For Döngüsü** ve **While Döngüsü**

````python
#Bir For döngüsü oluşturalım.
for i in range(1,11):
  print(i)
#For döngüsünün yazımı aslında çok basit:
#for .... in ..... :
#for, in ve iki nokta üst üste her zaman sabittir.
#Yukarıdaki i bir değişkendir ve for döngüsünde saymak için kullanılır.
# in yazısının sağındaki yazılansa i'nin içinde tek tek yazdığı büyük veridir.
#Bu büyük veri bazen bir liste bazen bir değişken olabilir.
#Yukarıdaki örnekte range dediğimiz şey bir aralıktır.
#range, 1 ile 10 arasındaki sayıları saymayı sağlıyor.
#Her zaman aralıkta yazılanın bir eksiğini verir.
````

# While Döngüsü
While döngüsü, for'dan farklı olarak bir koşulun oluşması durumunda çalışır. Bu koşul, sayıların dışında mantıksal koşulları da çalıştırır. En temelinde if ve for'un karışımı gibidir. Örneğin:

1. Eğer a'nın değeri 40'tan küçük olduğu sürece çalış,
2. Eğer parola Doğru olduğu müddetçe çalış,
3. Eğer kullanıcı_adı "Haşmet Dayı" olduğu sürece çalış.

While, her zaman bir koşula ihtiyaç duyar!

````python
#While'ı kullanıcı girişi olarak yazabilirim.
bilgi = input("Kullanıcı adınızı giriniz:")
while(bilgi != "Zeynep"):
  print("Kullanıcı adı yanlış!")
  bilgi = input("Kullanıcı adınızı giriniz:")
print("Kullanıcı adı doğru!")
````

````python
#For gibi kullanabilirim.
sayi1 = 0
while (sayi1<=100):
  print(sayi1)
  sayi1 += 1
````

````python
while(True):
  print("Yaz")
````

**NOT:** Asla sonsuz döngü kurmayın! Yoksa program döngüden asla çıkamaz!

````python
toplam = 0
while(True):
  print("Çıkmak için 0 yazınız.")
  bilgi = int(input("Lütfen bir sayı giriniz:"))
  toplam += bilgi #toplam = toplam + bilgi ile aynıdır.
  if bilgi == 0:
    break #Döngüyü kırmak demek
print("Toplanılan değer:", toplam)
````

**NOT:** While döngüsünü kırmak için "break" yani kırmak anlamına gelen komutu kullanırız. Eğer bir döngü veya if koşulunu pas geçip içi sadece boş kalsın istersek "pass" komutunu kullanırız.

````python
sayim = int(input("Lütfen sayi gir:"))

if sayim % 2 == 0:
  pass #If koşulunu atlıyor.
if sayim % 4 == 0:
  print ("sayı 4'e bölünebiliyor.")
````
