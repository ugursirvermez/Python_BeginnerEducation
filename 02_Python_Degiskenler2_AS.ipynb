#Kullanıcıdan Değer Alma
**Input Komutu:** Bu input komutu, kullanıcıdan veri almamızı sağlar.
Aldığı verileri metin değerinde saklar!<br>

Sayıları almak için değişkeni **int()** parantezine almamız gerekir!

```python
#Sayı değişkenlerini kullanıcıdan alma şu şekildedir:

yas = float(input("Lütfen yaşınızı giriniz:"))

#Yaşınız: 12
print("Yaşınız:"+str(yas)) # + işareti metinleri birleştirir. Dolayısıyla yanına sayı yazamazsınız.
print("Yaşınız:",int(yas)) # , işareti programlama nesnelerini yan yana yazabilmeyi sağlar. Sayı ve metin bir arada yazılabilir.
print("Yaşınız:"+str(int(yas))) # İç içe str int yapılabilir.
print(int(yas))
```

#Değişkenlerle İşlemler
String yani metin değişkenlerinin kendilerine özgü işlemleri, sayıların kendine özgü ayrı işlemleri vardır. İlk olarak sayı değişkenleri ile işlemlere başlayalım.

**NOT:** "print"'te + işareti aynı türden değerleri yan yana yazar veya toplar.
,(virgül) işareti farklı türden nesneleri ayrı olarak algılar ve yan yana yazabilmemizi sağlar.

```python
#Sayı değişkenini işleme sokma işlemi:
birinci_sayi = int(input("İlk sayıyı yazınız:"))
ikinci_sayi = int(input("İkinci sayıyı yazınız:"))
#NOT: Genelde bir işlem yapılıyorsa o işlemi ayrı bir değişkene yazarız.
#NOT-2: input'u int() parantezine alarak yazabiliriz. int(input(vs))
toplam = birinci_sayi + ikinci_sayi
cikarma = birinci_sayi - ikinci_sayi
carpma = birinci_sayi * ikinci_sayi #Çarpma işareti yıldızdır.
bolme = birinci_sayi / ikinci_sayi
print(birinci_sayi,"ile",ikinci_sayi," arasında 4 işlem şu şekildedir:")
print("Toplam:",toplam)
print("Çıkarma:",cikarma)
print("Çarpma:",carpma)
print("Bölme:", bolme)
```

#Değişkenlerle Sayısal İşlemler
Metin işlemlerinden farksızdır. Sadece önemli olan değişkenlerin int mi yoksa str mi olduğunu kontrol ederek işlem yapmamız gerekir.
int -> Integer -> Sayılar
str -> String -> Metin

```python
#Basit dört işlem yapmak
#input her zaman string yani metin türünde değerler alır. Sayıya çevirmek için int parantezini alırım.
sayi1 = int(input("Lütfen bir sayi giriniz:")) #Metini sayıya çevirmek icin input'u int parantezine alırım.
sayi2 = int(input("Lütfen bir sayi daha giriniz:"))
# toplam = 0
# cikarma = 0
toplam,cikarma,carpma,bolme = 0, 0, 0, 0 #Değiskenleri yan yana değerleri ile birlikte yazabilirim.
toplam = sayi1 + sayi2
cikarma = sayi1 - sayi2
carpma = sayi1 * sayi2
bolme = sayi1 / sayi2
print("Toplam:",toplam)
print("Çıkarma:",cikarma)
print("Carpma:",carpma)
print("Bölme:", bolme)
```

#ÖDEVLER
Bu haftanın ödevi oldukça basit. Geçen hafta yaptığınız şirket ödevine benzer bir ödev olacak. Ancak burada metinler değil sayılar kullancaksınız.

1. Kullanıcıdan **"Lütfen ilk notunuzu giriniz:"** şeklinde bir input isteyin. Bu input'un adı **"sayi1"** olacak.
2. Kullanıcıdan **"Lütfen ikinci notunuzu giriniz:"** şeklinde bir input isteyin. Bu input'un adı **"sayi2"** olacak.
3. Girilen iki sayiyi **"toplam""** adındaki değişkenin içine iki sayının toplama işlemini yaptırınız.
4. **"ortalama"** adındaki değişkenine de **"toplam"** değişkenindeki çıkan sayıyı da ikiye bölen ve **"ortalama"** değişkenini ekrana yazdırın.
5. **ÖZETLE;** Yukarıda şu ilemi yapıyoruz:
**(BirinciSayi + İkinciSayi)/2 = İki Notun Ortalaması**

```python
sayi1 = int(input("Birinci sayıyı giriniz:"))
sayi2 = int(input("İkinci sayıyı giriniz:"))

#Birinci Yöntem
toplam = sayi1 + sayi2
ortalama = toplam / 2

#İkinci Yöntem
ortalama = (sayi1 + sayi2) / 2

#print'te f komutu ile yazma
print(f"Birinci notum:{sayi1}, İkinci Notum:{sayi2}, Ortalaması ise {ortalama}'dır.")
```
