# While Döngüsü
While döngüsü, for'dan farklı olarak bir koşulun oluşması durumunda çalışır. Bu koşul, sayıların dışında mantıksal koşulları da çalıştırır. En temelinde if ve for'un karışımı gibidir. Örneğin:

1. Eğer a'nın değeri 40'tan küçük olduğu sürece çalış,
2. Eğer parola Doğru olduğu müddetçe çalış,
3. Eğer kullanıcı_adı "Haşmet Dayı" olduğu sürece çalış.

While, her zaman bir koşula ihtiyaç duyar!

```python
while(True):
  print("Muhammed")
```

**NOT:** Yukarıdaki örnek sonsuza kadar çalışır. <br> Fazla kullanmak sağlığa zararlıdır. Bilgisayarın sağlığına zararlıdır.

```python
sayi = 0
while(sayi<=10): #Sayi 10'a eşit ve 10'dan küçük olduğu sürece çalış
  print(sayi)
  sayi+=1 #Sayiya 1'er 1'er sayı ekle.
```

```python
bilgi = input("Kullanıcı Adınızı Yazınız:")
while(bilgi != "Haşmet Dayı"):
  print("Kullanıcı adı yanlış!")
  bilgi = input("Kullanıcı Adınızı Yazınız:")
print("Giriş Başarılı! Hoşgeldin Yüce Haşmet Dayı!")
```

**Sayı Tahmin Oyunu (Haşmet Dayı Edition)**

```python
import random

#Kullanıcı yaşını girecek.
# 1 - 10 yaş arası -> 1-20 arası rastgele sayı üretilecek.
# 10 - 20 yaş arası -> 1-50 arası rastgele sayı üretilecek.
# 20 - 50 yaş arası -> 1-100 arası rastgele sayı üretilecek.

print("Haşmet Dayının pizza dükkanına hoşgeldiniz!")
print("Pizzanızın içinden çıkabilecek böcek sayısını doğru tahmin ederseniz")
print("pizzanız bedava!")
yas = int(input("Lütfen önce yaşınızı giriniz:"))
rastgele_sayi = 0
if (yas >= 1 and  yas <= 10):
  rastgele_sayi = random.randint(1,20)
  print("1-20 arasında sayı tahmin edeceksiniz.")
if (yas >= 11 and  yas <= 20):
  rastgele_sayi = random.randint(1,50)
  print("1-50 arasında sayı tahmin edeceksiniz.")
if (yas >= 21 and  yas <= 50):
  rastgele_sayi = random.randint(1,100)
  print("1-100 arasında sayı tahmin edeceksiniz.")

tahmin_hakki = 5
tahmin = int(input("Lütfen tahmininizi giriniz:"))
while (tahmin != rastgele_sayi):
  tahmin_hakki -= 1
  print("Tahmininiz yanlış.",tahmin_hakki," hakkınız kaldı.")
  tahmin = int(input("Tekrar tahmnin et:"))
  if tahmin_hakki == 0:
    print("Tahmin hakkınız bitti!")
    print("Üretilen sayı",rastgele_sayi)
    break #While döngüsünü kırıyoruz.

  if (tahmin == rastgele_sayi):
    print("Tebrikler! sayıyı doğru bildiniz!")

print("Böcek sayısını tahmin edemediniz. Ücretiniz: 15TL")
```

```python
carpim = 1 #Sayıların toplanacağı değişken
eklenen_sayi = int(input("Lütfen sayıyı giriniz:"))
while(eklenen_sayi != 0):
  carpim *= eklenen_sayi #Girilen her bir sayıyı toplama ekle.
  print("çıkmak için 0'a basınız:")
  eklenen_sayi = int(input("Lütfen tekrar sayı giriniz:"))
print("Girdiğiniz sayıların çarpımı:", carpim," olarak görülmektedir.")
```

#For Döngüsü İle Yıldız Üçgeni Yapma


```python
for k in range(1,10,1):
  for i in range(k):
    print("*",end="")
  print()
for a in range(10,1,-1):
  for b in range(a):
    print("*",end="")
  print()
```
