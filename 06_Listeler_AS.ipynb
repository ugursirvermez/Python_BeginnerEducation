#Listeler
Değişkenler bir tane veri saklayan ve tek amaçla oluşturulan yapılardı. Listeler ise birden fazla veriyi içinde saklayan ve farklı türlerden verileri kendi içinde tutabilen yapılardır. <br>

Liste oluşturmak oldukça basittir. Öncelikle listenin adı verilir ve eşittir işareti konulur ve ardından köşeli parantezler [] arasına değerler yazılır.

````python
sehirler = ['Bursa', 'İstanbul', 'Ankara', 'İzmir']
haftanin_gunleri = ['Pazartesi', 'Sali', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
rastgele_liste = ["Elma", 12, "Bilinmez", "Çorba"]
print(sehirler)
print(haftanin_gunleri)
print(rastgele_liste)
````

````python
maker = ["Müfit Führer", "Zeynep KPop", "Ataberk Fener", "Yiğit Gman", "Uğur Kratos"]
print(maker[1])
````

**NOT:** Listelerde bir elemanı içinden çağırmak istiyorsak listenin adını yazıp yanına köşeli parantez [ ] açıp içerisine rakam yazıyoruz. Bu rakam 0'dan başlar ve listenin eleman sayısının 1 eksiğine kadar gider. Mesela haftanın günleri 7'yse listedeki elemanları 0 ve 6 arasında yazdırabilirim. <br>

Böylece ilk eleman 0'dır. Bu köşeli parantez içindeki değerlere **listenin indisi** denir.



````python
import random
listem = ["elma", "armut","ayva",124,True]
rastgele_indis = random.randint(0,len(listem)-1) #İndis sayısı eleman sayısından 1 eksik olmalı
print(listem[rastgele_indis])
````

#Listelere Input İle Eleman Ekleme ve Çıkarma
Input kullanarak listelere eleman ekleyip çıkarabiliriz. Bazı programlama dillerinde add (ekleme) veya append (listeye içine ekleme) anlamına gelen bir fonksiyon ile ekleme yapılabilir. <br>

Çıkarma işlemlerinde delete (silme) veya remove (listeden çıkarma) anlamında bu fonksiyonlar kullanılabilir. İkisinden örnek vermek gerekirse liste.append(sayi) veya liste.remove(3) gibi farklı şekillerde kullanılabilir.

````python
tatlilar = ["triliçe","kemal paşa", "sütlaç", "puding","baklava","pasta","kadayıf"]
tatli = input("Lütfen tatlı giriniz:")
tatlilar.append(tatli)
print(tatlilar)
#Listeden ekleneni silebilirim.
tatlilar.remove(tatli)
print("Silme İşlemi")
print(tatlilar)
silinecek = input("Listeden silinecek başka bir tatlı seçin:")
tatlilar.remove(silinecek)
print("Ekstra silmeden sonra:")
print(tatlilar)
````

#Listelerde Farklı İşlemler
Listeleri temizleme, eleman seçip aynı anda çıkarma, listeyi tersten yazdırma, liste eleman sayısı veya indis bilgisi alma gibi işlemler yapabiliriz.

````python
# Clear komutu -> Her şeyi sildi.
haftanin_gunleri = ['Pazartesi', 'Sali', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
haftanin_gunleri.clear()
print(haftanin_gunleri)
````

````python
#Count komutu -> Yazılan Elemani Sayar.
haftanin_gunleri = ['Pazartesi', 'Sali', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
gun = haftanin_gunleri.count('Sali')
print(gun)
````

````python
#Index komutu -> Yazılan elemanın kaçıncı index olduğunu yazar.
haftanin_gunleri = ['Pazartesi', 'Sali', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
print(haftanin_gunleri.index('Çarşamba'))
````

````python
#Pop komutu -> listenin içinden bahsedilen index'li elemanları alır yazar ve listeden çıkarır.
haftanin_gunleri = ['Pazartesi', 'Sali', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
gun = haftanin_gunleri.pop(3)
print(gun)
print(haftanin_gunleri)
````

````python
#Reverse komutu -> Listeyi Ters Çevirir
haftanin_gunleri = ['Pazartesi', 'Sali', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']
haftanin_gunleri.reverse()
print(haftanin_gunleri)
````

#Listelerde Arama Yapma ve Döngü Kullanımı
Listeler içerisinde belirli koşullar sağlandığında yazdırılacak veya işe yarayacak bilgileri alabilmek için listelerde arama yapılmalıdır. Listelerde arama yapmak için döngüler kullanılabilir.

````python
# 0 - 250 arasına 3'er 3'er sayıları yazdırma
ucerli_liste = []
onbesin_kati = []
for a in range(0,250,3):
  ucerli_liste.append(a)
print(ucerli_liste)
print("Toplam Eleman Sayısı",len(ucerli_liste))
# Bu listenin içinden 5'e bölünen sayıları ekrana yazdırma
for eleman in range(0, len(ucerli_liste)-1):
  if ucerli_liste[eleman] % 5 == 0:
    onbesin_kati.append(ucerli_liste[eleman])
print(onbesin_kati)
print("Toplam Eleman Sayısı",len(onbesin_kati))
````
