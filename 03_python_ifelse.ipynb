#Python'da IF-Else Koşulları
Python'da değişkenlerin içindeki değerlere göre işlem yapabilirim. Programlarımızda kullanıcıdan aldığımız değerlere göre koşullar oluşturup kullanıcının isteğine göre işlemler yapabilirim.

```python
#IF-Else Örnek Kullanımı
#Boolean Değişkeni sadece True (Doğru) veya False(Yanlış)
#değeri alan değişkenlerdi.
cinsiyet = True #Kadınsa True erkekse False olsun.
if cinsiyet == True: #Eğer cinsiyet kadınsa...
  print("Merhaba hanımefendi")
else: #değilse (kadın değilse)
  #İf veya else içinde tab ile boşluk bırakarak yazarım.
  print("Merhaba beyefendi")
print("Program bitti...")
```
