#Değişkenlerle Dört İşlem
Toplama, çıkarma, çarpma ve bölme işlemleri kolaylıkla yapılabilir.

Sayılar int yani integer değerlerle yazılır. Tamsayı anlamına gelir.

```python
sayi1 = int(input("Birinci sayiyi gir:"))
sayi2 = int(input("İkinci sayiyi gir:"))
#Dört İşlemi ismine göre değişkenlerde yazalım.
toplam = sayi1 + sayi2
cikarma = sayi1 - sayi2
carpma = sayi1 * sayi2 #Çarpma da yıldız kullanılır.
bolme = sayi1 / sayi2
print(f"Toplama:{toplam},Çıkarma:{cikarma},Çarpma:{carpma},Bölme:{bolme}")
```

```python
#Bir karenin ve bir dikdörtgenin alanını hesaplayınız.
#Karenin alanı: bir kenarı kendisiyle çarpmaktır.
#Dikdörtgenin alanı: kısa kenar ile uzun kenarın çarpımıdır.
#Kullanıcıdan 3 değişken iste bunlar:
#Karenin kenarı, dikdörtgenin kısa ve uzun kenarı olacak.

kkenari = int(input("Karenin kenarını giriniz:"))
dk_kisa = int(input("Dikdörtgenin kısa kenarını giriniz:"))
dk_uzun = int(input("Dikdörtgenin uzun kenarını giriniz:"))
kare_alan = kkenari * kkenari
dk_alan = dk_kisa * dk_uzun
print(kare_alan)
print(dk_alan)
```
