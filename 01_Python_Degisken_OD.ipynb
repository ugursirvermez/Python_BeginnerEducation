#DEĞİŞKENLER
Verileri saklamamızı sağlayan yapılardır. Diyelim ki kullanıcı bilgisayara bilgi girdi. Girilen bilgiyi ben programımda işlemek istersem o bilgiyi bir değişkenin içinde saklarım.

```python
#Bir değişken oluşturalım.
#Değişken adimi ve soyadimi saklasin
adsoyad = "Uğur Sırvermez"
yas = 27
print(adsoyad) #Değişkenler print edilmezse ekranda gözükmez!
print(yas)
```

#Değişkenlerin Özellikleri


1.   Değişkenler amacına göre isimlendirilir.
2.   İsim verilirken Türkçe karakter kullanılmaz.
3.   Değişkenler, sayı, metin, kesirli sayı, Doğru/Yanlış değerlerini alabilir.



```python
#Farklı değişken türleri
arabanin_ismi = "Tesla" #Metin -> String (str) demektir.
arabanin_motor_gucu = 400 # Sayı -> Integer (int) demektir.
arabanin_yuksekligi = 123.9 # Yükseklik -> Float (f) demektir.
arabanin_genisligi = 95.9 #Ondalıklı Sayı -> Double (d) demektir.
araba_calisiyor_mu = True # Doğru/Yanlış -> Boolean (bool) demektir.
araba_pahali_mi = False # Doğru/Yanlış -> Boolean (bool) demektir.
print(arabanin_ismi, arabanin_motor_gucu, arabanin_yuksekligi, arabanin_genisligi, araba_calisiyor_mu, araba_pahali_mi)
```
