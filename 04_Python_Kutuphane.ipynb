#Kutuphaneler

---
Önceden hazırlanmış, bir çok özel işlemi içinde barındıran kullanıma hazır yapılardır. Python'da sayısız kütüphane vardır. Örneğin:

*   Matematiksel İşlemler için **math**
*   Rastgele içerik ve sayı üretmek için **random**
*   İşletim sistemi dosyalarına erişmek için **os** kütüphanesi
*   Oyun geliştirmek için **turtle** veya **pygame** kütüphaneleri
*   Arayüzlü program için **tkinter** kütüphanesi

##Peki Kütüphaneler Nasıl Eklenir?

Kütüphaneleri eklemek için genelde kullanılan yöntem şu şekildedir:<br>
import kütüphane_adi <br>

Örneğin;  
import random





```python
#Kütüphaneler her zaman projenin en başına eklenir.
import random #Rastgele içerik veya sayı üretmek için kullanılır.

rastgele_sayi = random.randint(1,11) #1-11 arasında rastgele sayi seç.
print(rastgele_sayi)
if rastgele_sayi == 1:
  print("Uğur")
```

```python
kullanici_adi = input("Lütfen kullanıcı adınızı yazınız:")
parola = input("Lütfen parolanızı yazınız:")
if kullanici_adi == "Haşmet Dayı":
  if parola == "Minecraft Sıkıcı":
    print("Tebrikler başarılı giriş!")
  else:
    print("Şifre Yanlış!")
else:
  print("Kullanıcı Adı Yanlış!")
```

##Kütüphanelerde Kısaltma
Bazen kütüphanelerin ismi uzundur. Biz o isimleri kısaltmak için
**as** takısı kullanarak kütüphane isimlerini kısaltırız.

```python
import random as rd #Kısaltma yapılabiliyor.
import math as m #m'yi kısalttık.

sayi1 = rd.randint(1, 10) #1-10 arasında rastgele sayıyı değişkene atabiliyoruz.

print(sayi1)

#Matematiksel Fonksiyonlar
print(m.factorial(sayi1)) #Faktöriyel
print(m.comb(sayi1, 2)) #Kombinasyon (iki değer alır) (sayi1, sayi2)
print(m.perm(sayi1, 2)) #Permitasyon
print(m.sqrt(sayi1)) #Karekök alma
print(m.floor(sayi1)) #Yuvarlama
print(m.pow(sayi1, 4)) #İstediğin sayının Kökkuvvetini alma
print(m.log(sayi1, 3)) #İstediğin tabandan logaritma
print(m.pi) #Pi sayısı
print(m.e) # e sayısı
```
