#Değişken Oluştururken Dikkat!

1.   Değişken ismini değişkenin amacına göre içinde barındırdığı bilgiye göre vermemiz gerekir.
2.   Değişken isimlerini oluştururken Türkçe karakter kullanmıyorum. (**Türkçe karakterler: ı,ç,ö,ü,ş,ğ,İ**)
3.   Değişken oluştururken isimler arasına boşluk bırakmadan yazarız. <br>
  Ad Soyad (**Yanlış**)<br>
  Ad_Soyad (**Doğru**)



````python
#Değişken Türleri
#Bir Arabanın Özellikleri
araba_marka = "Tesla" #Metin -> String (str) demektir.
araba_hiz = 4000 #Sayi -> Integer (int) demektir.
arabanin_yuksekligi = 139.9 #Yukseklik -> Float (f) demektir.
arabanin_genisligi = 250,3 # Genişliği -> Double (d) demektir.
arabanin_uzunlugu = 390.2 # Float kullanilir.
araba_calisiyor_mu = True #Doğru/Yanlış -> Boolean (bool) demektir.
araba_pahali_mi = False #Doğru = True, Yanlış = False
#Tek print'te ama alt alta bütün değişkenlerin değerlerini nasıl yazarım?
print(araba_marka+"\n"+str(araba_hiz)+"\n"+str(arabanin_yuksekligi)
      +"\n"+str(arabanin_genisligi)+"\n"+str(arabanin_uzunlugu)
      +"\n"+str(araba_calisiyor_mu)+"\n"+str(araba_pahali_mi))
````
