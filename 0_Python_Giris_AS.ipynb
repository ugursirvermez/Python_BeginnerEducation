#Python Giriş Eğitimi
Bu derste Print fonksiyonu ile ekrana yazı yazdırma, print fonksiyonunun temel özellikleri yer alıyor.

```python
print("HELLO WORLD!")
```

```python
print('HELLO WORLD!')
```

```python
print(8)
print(12)
print(8+12) #Toplama
print(8-12) #Cikarma
print(8*12) #Carpma
print(8/12) #Bolme

print(9/3)
```

**String:** Metin yazılı değişkenlerine verilen isimdir. Çift veya tek tırnak içerisinde yazılır. Kısaltması str'dir. *Örnek: print("Selam")*

**Integer:** Sayı değişkenlerine verilen isimdir. Sadece rakam veya sayı yazılır. Kısaltması int'tir. *Örnek: print(1395)*

```python
#print(8+""+12) TypeError: unsupported operand type(s) for +: 'int' and 'str'
print(str(8)+""+str(12))
#print(int("matematik")) ValueError: invalid literal for int() with base 10: 'matematik'
```

**NOT:** Sayıları integer'den string yani metine **dönüştürebilirim** ancak metinleri sayıya **çeviremem!**

```python
print(len("MATEMATİK"))
```

**len()** -> length türkçesi uzunluk demektir. Bir metnin kaç karakter olduğunu bize söyler.

**Görev:** Atatürk'ün dediği gibi "Egemenlik kayıtsız şartsız milletindir."

```python
print("Atatürk\'ün dediği gibi \"Egemenlik kayıtsız şartsız milletindir.\"")
print('''Atatürk'ün dediği gibi "Egemenlik kayıtsız şartsız milletindir."''')
```

**NOT:** Başına ve sonuna 3 tane tek tırnak koyarsak içinde özel karakterleri çalıştırabiliriz. (?,"", '')

**Kaçış Dizisi(\\):** Kaçış dizisi özel karakter tanımlamak için kullanılan bir kavramdır.

```python
print((8-12),(8+12),(8*12),(8/12))
print((8-12)+(8+12)+(8*12)+(8/12))
```

```python
print("Bugün","Hava","Çok","Güzel")
print("Bugün"+"Hava"+"Çok"+"Güzel")
```

**NOT:** Print komutunda sayılarda virgül ile yan yana işlemler sırasıyla yaptırılabilir. Ancak metinlerde virgül ve artı işareti ile yan yana yazdırılabilir.

```python
print("Bugün","Hava\n","Çok","Güzel")
```

**\n:** Alt satıra yazıyı indir.
**\r:** Kendisinden öncekileri siler.
**\t:** Kendisinden sonraya 4 boşluk bırakır.

```python
print("Bugün\r","Hava","Çok","Güzel")
print("Bugün","Hava","Çok\r","Güzel")
```

```python
print("Bugün\t","Hava","Çok","Güzel")
print("Bugün","Hava","Çok\t","Güzel")
```
